## Clase 01 - Introducción a Señales y Sistemas

### ¿Qué es una señal?

Una señal es la representación matemática de una o más variables independientes, en términos de una función. Algunos ejemplos de señales, en el contexto de la Ingeniería en Telecomunicaciones, pueden ser la corriente o el voltaje en el tiempo (En este caso, el tiempo es la variable independiente).

Las señales se pueden clasificar como señales continuas o señales discretas.

### Señal continua

Las señales que existen en la naturaleza son de naturaleza continua, esto es, tienen un valor continuo en el rango de −∞ hasta +∞.

Una señal continua representa una serie de datos que describen un fenómeno dado a lo largo de una variable independiente que es continua. Cuando la variable independiente es el tiempo **t**, se conocen como "Señales continuas en el tiempo". Es el caso de la temperatura **T** que sentimos en un día cualquiera **T(t)**, la cual varía lentamente y de forma continua. De igual manera, la voz **v** que transmitimos al hablar tiene una forma continua en el tiempo **v(t)**.
Sin embargo, la variable independiente no siempre es el tiempo. Por ejemplo, en el caso de un sonido que se emite a un volumen y frecuencias fijas, el nivel de volumen percibido **v** puede depender de la distancia **d** de quien lo escucha **v(d)**.

### Señal discreta
Representan una serie de datos que no son continuos y que, mediante un proceso de muestreo, describen un fenómeno dado a ciertos intervalos. En este caso, la señal es muestreada cada cierto tiempo t (intervalo de muestreo) y obtiene valores específicos. Si la señal discreta puede tomar sólo algunos valores específicos, se dice que es una señal digital.

**Funciones utilizadas:**

* np.arange(0, 10, dt): Genera un array de valores en el rango de 0 a 10 (excluyendo 10) con un incremento de dt, que se utiliza para representar los momentos de muestreo en el tiempo.

* np.sin(np.pi * t) * np.exp(-0.5 * t): Calcula la señal discreta x a partir de una función sinusoidal y una función exponencial, utilizando los valores discretos de tiempo t.

* plt.stem(t, x, label='x(t) = Sin(PI*t)*e^(-0.5*t)'): Crea un diagrama de tallo y hoja (stem plot) de la señal discreta x en función de los momentos de muestreo t.

* axs.set_xlabel("Tiempo"): Establece la etiqueta del eje x.

* axs.set_ylabel("Amplitud"): Establece la etiqueta del eje y.

* plt.legend(loc='upper right'): Agrega una leyenda en la esquina superior derecha de la gráfica.

* plt.show(): Muestra la gráfica generada.

```python
dt = 0.2 # Intervalo de muestreo
Fs = 1 / dt # frecuencia de muestreo
t = np.arange(0, 10, dt)
# Señal a graficar x(t) = Sin(PI*t)*e^(-0.5*t)
x = np.sin(np.pi * t) * np.exp (-0.5 * t)
fig, axs = plt.subplots()
axs.set_title("Señal discreta en el tiempo")
markerline, stemlines, baseline = plt.stem(t, x, label='x(t) = Sin(PI*t)*e^(-0.5*t)')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```

#Señales de soporte compacto (De duración finita)

Son señales de duración finita, sólo tienen valores diferentes de cero en un intervalo definido.

El soporte es el subconjunto de tiempo en donde se presentan los valores diferentes de cero en la señal. Este subconjunto debe ser conexo y el mínimo posible.

**Funciones utilizadas:**

* np.linspace(0, 1, num=1000): Genera un array de valores equidistantes en el rango de 0 a 1 con un total de 1000 valores. Estos valores representan los momentos de tiempo t.

* np.sin(2 * np.pi * 10 * t): Calcula una onda sinusoidal de frecuencia 10 Hz modulada por los valores de tiempo t.

* (t >= 0.1) * (t <= 0.3): Crea una máscara booleana que es verdadera entre 0.1 y 0.3 segundos y falsa en otros lugares. Esto limita la señal a tener valor no nulo solo en ese intervalo.

* plt.plot(t, x): Grafica la señal x en función de los valores de tiempo t.

* plt.xlabel('Tiempo (s)'): Establece la etiqueta del eje x.

* plt.ylabel('Amplitud'): Establece la etiqueta del eje y.

* plt.title('Señal de Soporte Compacto'): Establece el título de la gráfica.

* plt.grid(True): Agrega una cuadrícula a la gráfica.

* plt.show(): Muestra la gráfica generada.

```python
# Definir la señal de soporte compacto entre t=0.1 y t=0.3
t = np.linspace(0, 1, num=1000)
x = np.sin(2 * np.pi * 10 * t) * (t >= 0.1) * (t <= 0.3)

# Graficar la señal
plt.plot(t, x)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal de Soporte Compacto')
plt.grid(True)
plt.show()
```

#Señal acotada en valor

Son señales que **solo toman valores finitos**, su rango de valores está acotado (Tiene un límite).

En el siguiente código, la señal $sin(10\pi t)$ sólo puede tomar valores por debajo de 0.3.

```python
# Definir la señal
t = np.linspace(0, 0.2, num=1000)
x = np.sin(2 * np.pi * 5 * t)

# Acotar la señal (Recortar)
x = np.clip(x, -1, 0.3)

# Graficar la señal
plt.plot(t, x)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal acotada')
plt.grid(True)
plt.show()
```

# Señales periódicas

Una señal periódica continua $x(t)$ es aquella que tiene un valor $T$ para el cual:

$$x(t)=x(t+T)$$

Es decir que, al moverse un tiempo $T$ la señal seguirá siendo la misma. En este caso, se trata de una señal periódica con periodo $T$. Este tipo de señales son muy comunes en diversos contextos.

## Señal periódica continua

Si $x(t)$ es periódica con periodo $T$, entonces se cumple que $x(t)=x(t+mT)$, para toda $t$ y para cualquier entero $m$. Esto quiere decir que también es periódica para periodos $2T$, $3T$, $4T$,...

El *periodo fundamental* de la señal es $T_0$ de $x(t)$ es el valor positivo más pequeño de $T$ para el cual $x(t)=x(t+mT)$ se satisface. Esto es válido, excepto cuanto $x(t)$ es una constante (el periodo fundamental es indefinido en este caso).

```python
#La variable independiente t se tomará en el rango [0,10] y se graficará cada 0.01 segundos.
t = np.arange(0, 10, 0.01)

# Señal a graficar x(t) = Sin(2*PI*t)
x = np.sin( 2*np.pi * t)

# Configuración de la gráfica en particular que se va a dibujar
fig, axs = plt.subplots()
axs.set_title("Señal periódica en el tiempo, con periodo w/(2PI)")
axs.plot(t, x, color='C0', linewidth=2, label='x(t) = Sin(2*PI*t)')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```

## Señal periódica discreta

De forma similar, una señal periódica discreta se define como:

$$x[n]=x[n+N]$$

Para cualquier valor de $n$. Esta señal también es periódica para periodos de $2N$; $3N$, etc. En este caso, el periodo fundamental $N_0$ será el valor más pequeño de $N$ para el que se cumple la ecuación de periodicidad.

¿Cuál será el periodo de la siguiente señal?

```python
dt = 0.2 # Intervalo de muestreo
Fs = 1 / dt # frecuencia de muestreo
t = np.arange(-2.5, 2.5, dt)
# Señal a graficar x(t) = Sin(2*PI*t)
x = np.sin (2 * np.pi * t)
fig, axs = plt.subplots()
axs.set_title("Señal de valor discreto")
#La función around redondea el dato a su valor más cercano, según el número de decimales indicado
markerline, stemlines, baseline = plt.stem(t, np.around(x,0), label='x(t) = Sin(2*PI*n)')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```

## Señal exponencial

La *Señal* continua *exponencial compleja* es de la forma:

$$x(t)=Ce^{at}$$

Donde $C$ y $a$ en general son números complejos. Estos valores le dan diversas formas a la exponencial compleja.

### Señales exponenciales reales

Cuando $C$ y $a$ son valores reales, la forma de la señal dependerá de si el parámetro $a$ es positivo o negativo.

```python
#La variable independiente t se tomará en el rango [-2.5,2.5] y se graficará cada 0.01 segundos.
t = np.arange(-2.5, 2.5, 0.01)

# Señal a graficar x(t) = e^(-0.5*t)
x1 = np.exp (-0.5 * t)

# Señal a graficar x(t) = e^(-0.5*t)
x2 = np.exp  (0.5 * t)

# Configuración de la gráfica en particular que se va a dibujar
fig, axs = plt.subplots()
axs.set_title("Señal exponencial real")
axs.plot(t, x1, color='C0', linewidth=2, label = 'x(t) = e^(-0.5*t)')
axs.plot(t, x2, color='C1', linewidth=2, label = 'x(t) = e^( 0.5*t)')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```

### Señal periódica exponencial compleja

Considerando el campo imaginario, se trata de una señal de tipo:

$$x(t)=e^{j\omega_0t}$$

Para que sea periódica, se debe tener en cuenta que:

$$e^{j\omega_0t}=e^{j\omega_0(t+T)}$$

$$e^{j\omega_0(t+T)}=e^{j\omega_0t}e^{j\omega_0T}$$

Para que sea periódica es necesario entonces que:

$$e^{j\omega_0T}=1$$

En este caso, si $w_0$ es 1, esta señal es periódica para cualquier $T$. Para $w_0\neq0$, el periodo fundamental $T_0$ se cumple cuando:

$$T_0=\frac{2\pi}{\omega}$$

Por lo tanto, las señales $x(t)=e^{j\omega_0t}$ y $x(t)=e^{-j\omega_0t}$ tienen el mismo periodo fundamental.

### Señal periódica senoidal

Una señal muy relacionada con la exponencial compleja es la senoidal, de la forma:

$$x(t)=A\cos({\omega_0t}+\phi)$$

Es importante tener en cuenta que la frecuencia angular $\omega$ se expresa en rad/s y que se relaciona con la frecuencia $f$ expresada en ciclos por segundo (o hertz, Hz) de la forma:

$$\omega_0=2{\pi}f_0$$

```python
#Se define la frecuencia angular w = 2*PI
w = 2*np.pi

#La variable independiente t se tomará en el rango [0,2*PI/w] y se graficará cada 0.01 segundos.
t = np.arange(0, 2*np.pi/w, 0.01)

#Se define el ángulo de desfase, este debe estar en radianes o convertirse posteriormente a radianes
ang = 0

# Señal a graficar x(t) = Sin(2*PI*t)
x = np.cos( 2*np.pi * t + np.radians(ang))

# Configuración de la gráfica en particular que se va a dibujar
fig, axs = plt.subplots()
axs.set_title("Señal periódica en el tiempo, con periodo w/(2PI)")
axs.plot(t, x, color='C0', linewidth=2, label='x(t) = Cos(2*PI*t)')
axs.set_xlabel("Tiempo")
axs.set_ylabel("Amplitud")
plt.legend(loc='upper right')
plt.show()
```

Las señales *exponencial compleja* y *senoidal* están relacionadas por la ecuación de Euler:

$$e^{j\omega_0t}=\cos{\omega_0t}+j\sin{\omega_0t}$$

La señal senoidal puede escribirse en términos de exponenciales complejas de la forma:

$$A\cos({\omega_0t}+\phi)=\frac{A}{2}e^{j\phi}e^{j\omega_0t}+\frac{A}{2}e^{-j\phi}e^{-j\omega_0t}$$

Y también, en términos de la señal exponencial compleja, de la forma:

$$A\cos({\omega_0t}+\phi)=A\Re\{e^{(j\omega_0t+\phi)}\}$$

$$A\sin({\omega_0t}+\phi)=A\Im\{e^{(j\omega_0t+\phi)}\}$$
